#### Load move_review_dataset.csv  for testing

```python
# import required libraries
import pandas as pd
import numpy as np
```

```python
move_review_df = pd.read_csv("move_review_dataset.csv",encoding='utf-8',compression='gzip')
move_review_df
```

```python
move_review_df.sort_values(by='year', ascending=True)
```

```python
# Null check
move_review_df.isnull().sum()
```

```python
# Some rows has null title ie invalid row,with out title no movie exits
# drop those null values
move_review_df.dropna(inplace=True)
```

```python
# fillna
move_review_df["genres"].fillna("No genres",inplace=True)
```

```python
# need to convert year string to numeric datatype to run queries
move_review_df.year = pd.to_numeric(move_review_df.year, errors='coerce').fillna(0).astype(np.int64)
```

```python
move_review_df["title"]
```

```python
# convert movie titles to locase case
#move_review_df["title"]=move_review_df["title"].apply(lambda x: x.lower())

# remove spaces like trim 
#move_review_df["title"]=move_review_df["title"].apply(lambda x: x.strip())
```

```python
# convert movie genres to locase case
move_review_df["genres"]=move_review_df["genres"].apply(lambda x: x.lower())

# remove spaces genres trim 
move_review_df["genres"]=move_review_df["genres"].apply(lambda x: x.strip())
```

```python
#convert dataframe to json and return json response
def moveiesByYear(year):
    movieList=move_review_df.query('year == @year')
    json=movieList.to_json(orient = "records")
    return json 
```

```python
moviesList=moveiesByYear(2018)
moviesList
```

```python
def moveieByName(name):
    name=name.strip().lower()
    movieList= move_review_df[move_review_df["title"].str.contains(name)]
    json=movieList.to_json(orient = "records")
    return json 
```

```python
moviesListByName=moveieByName('   fantastic  ')
moviesListByName
```

```python
def moveieByGenres(genres):
    genres=genres.strip().lower()
    movieList= move_review_df[move_review_df["genres"].str.contains(genres)]
    json=movieList.to_json(orient = "records")
    return json 
```

```python
moviesListByGenres=moveieByGenres('crime , dram')
moviesListByGenres
```

```python
def moveieByGenresAndYear(genres,year):
    genres=genres.strip().lower()
    genres=genres.replace("and",",")
    movieList= move_review_df[(move_review_df["genres"].str.contains(genres)) & (move_review_df["year"] == year)]
    json=movieList.to_json(orient = "records")
    return json 
```

```python
moviesListByGenresYear=moveieByGenresAndYear('crime , dram',2018)
moviesListByGenresYear
```
